```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Plotagem

```python
# Plotagem b√°sica: plot
dados = pd.Series(np.random.randn(15), index=pd.date_range('1/1/2000', periods=15))
dados 
```

```python
dados.plot()
```

```python
# No DataFrame, plot()√© conveniente plotar todas as colunas com r√≥tulos:
df = pd.DataFrame(np.random.randn(15,4), index=dados.index, columns=list('ABCD'))
df.cumsum()
plt.figure()
df.plot()
```

```python
# Voc√™ pode representar uma coluna em compara√ß√£o com outra usando as palavras x- ychave e em plot():
df = pd.DataFrame(np.random.randn(1000, 2), columns=['A', 'B']).cumsum()
df['A'] = pd.Series(list(range(len(df))))
df.plot(x='A', y='B')
```

```python
# Chamar o plot.bar()m√©todo de DataFrame produz um gr√°fico de v√°rias barras:
df = pd.DataFrame(np.random.rand(10, 4), columns=['a', 'b', 'c', 'd'])
df.plot.bar()
```

```python
# Para produzir um gr√°fico de barras empilhadas, passe stacked=True:
df.plot.bar(stacked=True)
```

```python
# Para obter gr√°ficos de barra horizontal, use o barhm√©todo:
df.plot.barh(stacked=True)
```

### Histogramas

```python
df = pd.DataFrame({
    "a": np.random.randn(1000) + 1,
    "b": np.random.randn(1000),
    "c": np.random.randn(1000) - 1,
},columns=["a", "b", "c"],)
plt.figure()
df.plot.hist(alpha=0.5)
```
